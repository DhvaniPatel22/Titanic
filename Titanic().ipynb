```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
```

```python
td.Sex_encoded=le.fit_transform(td.Sex)
print(td.Sex_encoded)
td.Embarked_encoded=le.fit_transform(td.Embarked)
print(td.Embarked_encoded)
```

```python
x = list(zip(td.Age,td.SibSp,td.Sex_encoded,td.Embarked_encoded,td.Parch,td.Fare))
print(x)
```

```python
y=td["Survived"]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2,random_state=109) 
```

```python
from sklearn.linear_model import LogisticRegression

logistic_reg = LogisticRegression()
logistic_reg.fit(x_train, y_train)
```

```python
from sklearn import metrics 
y_pred = logistic_reg.predict(x_test)

print("Accuracy:",metrics.accuracy_score(y_test, y_pred))

```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()
clf = clf.fit(x_train,y_train)
y_pred = clf.predict(x_test)
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
#Import svm model
from sklearn import svm

#Create a svm Classifier
clf = svm.SVC(kernel='linear') # Linear Kernel

#Train the model using the training sets
clf.fit(x_train, y_train)

#Predict the response for test dataset
y_pred = clf.predict(x_test)
svm=print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
svm
```

```python
from sklearn.naive_bayes import GaussianNB

#Create a Gaussian Classifier
gnb = GaussianNB()

gnb.fit(x_train, y_train)
y_pred = gnb.predict(x_test)
gn=print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
gn
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 3) 
knn.fit(x_train, y_train)  
y_pred = knn.predict(x_test)  
knn=print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
knn
```
